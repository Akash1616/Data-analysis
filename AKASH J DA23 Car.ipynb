# Data Preprocessing


## Missing value treatment
- For the predictor variables, we will replace the missing values in each column with its median.

```python
df = df_car.copy()

# Group by car maker and mode, calculate the mean, and fill missing values
df['City_Mileage_km_litre'] = df.groupby(['Make', 'Model'])['City_Mileage_km_litre'].transform(lambda x: x.fillna(x.mean()))
df['Highway_Mileage_km_litre'] = df.groupby(['Make', 'Model'])['Highway_Mileage_km_litre'].transform(lambda x: x.fillna(x.mean()))

# Replace 'NA kg' values in 'Gross_Vehicle_Weight' with the mean of the column
gross_vehicle_weight_numeric = pd.to_numeric(df['Gross_Vehicle_Weight'], errors='coerce')
mean_gross_vehicle_weight = gross_vehicle_weight_numeric.mean()
df['Gross_Vehicle_Weight'].replace('NA kg', mean_gross_vehicle_weight, inplace=True)

# Iterate over numeric, date, and categorical columns
for col_type in ['number', 'datetime', 'object']:
    # Identify columns of the current type
    columns = df.select_dtypes(include=[col_type]).columns
    
    # Fill missing values in selected columns based on column type
    if col_type == 'number':
        df.loc[:, columns] = df.loc[:, columns].fillna(df.loc[:, columns].mean())
    elif col_type == 'datetime':
        df[columns] = df[columns].apply(lambda x: x.fillna(x.mode()[0]))
    elif col_type == 'object':
        df[columns] = df[columns].apply(lambda x: x.fillna(x.mode()[0]))

```

## Separating Categorical and Numerical Variables

```python
def separate_data_types(df):
    categorical = []
    continuous = []
    for column in df.columns:
        if df[column].dtype == 'O':  # 'O' indicates object (categorical)
            categorical.append(column)
        elif df[column].dtype in ['int64', 'float64']:  # Numeric types (continuous)
            continuous.append(column)
    return categorical, continuous
```

```python
categorical_columns, numeric_columns = separate_data_types(df)
# Displaying the results using tabulate
table_df = tabulate({"Categorical": categorical_columns, "Continuous": numeric_columns}, headers="keys")

# Printing the tables
print("DataFrame: df")
print(table_df)
```

### **Task 3** 
Compare the performance of cars based on factors such as displacement, cylinders, and
valves per cylinder, providing valuable insights for marketing and product development.

```python

selected_columns = ['Make', 'Displacement', 'Cylinders', 'Valves_Per_Cylinder']

# Create a new DataFrame with selected columns
performance_df = df[selected_columns].copy()

# Convert 'Displacement' column to numeric (handling non-numeric values)
performance_df['Displacement'] = pd.to_numeric(performance_df['Displacement'], errors='coerce')

# Group by 'Make' and calculate mean, median, std, and coefficient of variation
performance_stats = performance_df.groupby('Make').agg({
    'Displacement': ['mean', 'median', 'std'],
    'Cylinders': ['mean', 'median', 'std'],
    'Valves_Per_Cylinder': ['mean', 'median', 'std']
}).round(2)

# Calculate coefficient of variation separately
cv_displacement = (performance_stats['Displacement']['std'] / performance_stats['Displacement']['mean']).round(2)
cv_cylinders = (performance_stats['Cylinders']['std'] / performance_stats['Cylinders']['mean']).round(2)
cv_valves = (performance_stats['Valves_Per_Cylinder']['std'] / performance_stats['Valves_Per_Cylinder']['mean']).round(2)

# Add coefficient of variation to the DataFrame
performance_stats[('Displacement', 'cv')] = cv_displacement
performance_stats[('Cylinders', 'cv')] = cv_cylinders
performance_stats[('Valves_Per_Cylinder', 'cv')] = cv_valves

# Display statistics
performance_stats
```

```python
# Box plot for 'Displacement' distribution for each manufacturer
plt.figure(figsize=(12, 8))
sns.boxplot(x='Make', y='Displacement', data=performance_df)
plt.title('Displacement Distribution by Manufacturer')
plt.xlabel('Manufacturer')
plt.ylabel('Displacement')

# Rotate x-axis labels by 90 degrees
plt.xticks(rotation=90)

plt.show()
```

```python
# Convert 'Seating_Capacity' to numeric, handle errors with 'coerce'
df['Seating_Capacity'] = pd.to_numeric(df['Seating_Capacity'], errors='coerce')

# Calculate average seating capacity for each car model
average_seating_capacity = df.groupby('Make')['Seating_Capacity'].mean().round(2)

# Count occurrences of central locking for each car model
central_locking_counts = df.groupby('Make')['Central_Locking'].value_counts().unstack().fillna(0)

# Count occurrences of child safety locks for each car model
child_safety_locks_counts = df.groupby('Make')['Child_Safety_Locks'].value_counts().unstack().fillna(0)

# Visualize average seating capacity
plt.figure(figsize=(12, 6))
average_seating_capacity.plot(kind='bar', color='skyblue')
plt.title('Average Seating Capacity by Car Model')
plt.xlabel('Car Model')
plt.ylabel('Average Seating Capacity')
plt.xticks(rotation=90, ha='right')
plt.show()

# Visualize counts of central locking
plt.figure(figsize=(12, 6))
central_locking_counts.plot(kind='bar', stacked=True, color=['skyblue', 'lightcoral'])
plt.title('Central Locking Counts by Car Model')
plt.xlabel('Car Model')
plt.ylabel('Count')
plt.xticks(rotation=90, ha='right')
plt.legend(title='Central Locking', loc='upper right')
plt.show()

# Visualize counts of child safety locks
plt.figure(figsize=(12, 6))
child_safety_locks_counts.plot(kind='bar', stacked=True, color=['skyblue', 'lightcoral'])
plt.title('Child Safety Locks Counts by Car Model')
plt.xlabel('Car Model')
plt.ylabel('Count')
plt.xticks(rotation=90, ha='right')
plt.legend(title='Child Safety Locks', loc='upper right')
plt.show()
```
