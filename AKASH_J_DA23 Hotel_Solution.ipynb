### Data Dictionary

1. **hotel** - Type of hotel (e.g., Resort Hotel).
2. **is_canceled** - Binary variable indicating whether the booking was canceled (1) or not (0).
3. **lead_time** - Number of days between booking and arrival.
4. **arrival_date_year** - Year of arrival date.
5. **arrival_date_month** - Month of arrival date.
6. **arrival_date_week_number** - Week number of year for arrival date.
7. **arrival_date_day_of_month** - Day of the month of arrival date.
8. **stays_in_weekend_nights** - Number of weekend nights (Saturday or Sunday) the guest stayed.
9. **stays_in_week_nights** - Number of weeknights (Monday to Friday) the guest stayed.
10. **adults** - Number of adults.
11. **children** - Number of children.
12. **babies** - Number of babies.
13. **meal** - Type of meal booked (e.g., BB for Bed & Breakfast).
14. **country** - Country of origin.
15. **market_segment** - Market segment designation (e.g., Online TA for Travel Agents).
16. **distribution_channel** - Booking distribution channel (e.g., TA/TO for Travel Agents/Tour Operators).
17. **is_repeated_guest** - Binary variable indicating whether the guest is a repeated guest (1) or not (0).
18. **previous_cancellations** - Number of previous bookings that were canceled by the customer.
19. **previous_bookings_not_canceled** - Number of previous bookings that were not canceled by the customer.
20. **reserved_room_type** - Code of room type reserved.
21. **assigned_room_type** - Code of room type assigned.
22. **booking_changes** - Number of changes made to the booking.
23. **deposit_type** - Type of deposit made (e.g., No Deposit).
24. **agent** - ID of the travel agency that made the booking.
25. **company** - ID of the company/entity that made the booking or responsible for paying.
26. **days_in_waiting_list** - Number of days the booking was on the waiting list.
27. **customer_type** - Type of booking (e.g., Transient).
28. **adr** - Average Daily Rate, calculated by dividing the sum of all lodging transactions by the total number of staying nights.
29. **required_car_parking_spaces** - Number of car parking spaces required by the guest.
30. **total_of_special_requests** - Number of special requests made by the guest.
31. **reservation_status** - Reservation last status (e.g., Check-Out).
32. **reservation_status_date** - Date when the last status was set.

## Task 2.3. How do booking patterns vary by month, week, and day of the week?

```python
# Convert arrival_date_year, arrival_date_month, and arrival_date_day_of_month to strings
df_hotel['arrival_date_year'] = df_hotel['arrival_date_year'].astype(str)
df_hotel['arrival_date_month'] = df_hotel['arrival_date_month'].astype(str)
df_hotel['arrival_date_day_of_month'] = df_hotel['arrival_date_day_of_month'].astype(str)

# Booking Patterns by Month
plt.figure(figsize=(12, 6))
sns.countplot(x='arrival_date_month', data=df_hotel, order=df_hotel['arrival_date_month'].value_counts().index, palette='viridis')
plt.title('Booking Patterns by Month')
plt.xlabel('Month')
plt.ylabel('Number of Bookings')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Booking Patterns by Week
plt.figure(figsize=(12, 6))
sns.countplot(x='arrival_date_week_number', data=df_hotel, palette='magma')
plt.title('Booking Patterns by Week')
plt.xlabel('Week Number')
plt.ylabel('Number of Bookings')
plt.tight_layout()
plt.show()

# Combine year, month, and day columns to create a datetime column
df_hotel['arrival_date'] = pd.to_datetime(df_hotel['arrival_date_year'] + '-' + df_hotel['arrival_date_month'] + '-' + df_hotel['arrival_date_day_of_month'], errors='coerce')

# Extract day of the week from the datetime column
df_hotel['day_of_week'] = df_hotel['arrival_date'].dt.day_name()

# Booking Patterns by Day of the Week
plt.figure(figsize=(10, 6))
sns.countplot(x='day_of_week', data=df_hotel, order=['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'], palette='rocket')
plt.title('Booking Patterns by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Number of Bookings')
plt.tight_layout()
plt.show()
```

**Interpretations:**
* The highest number of bookings by month occurs in July and August. For a week, it's the 18th week, and for a day of the week, it's Saturday.

### **Task 3: Booking Cancellation Analysis**

## Task 3.2. Are there specific trends or patterns in cancellations over time?

```python
# Convert reservation_status_date to datetime format
df_hotel['reservation_status_date'] = pd.to_datetime(df_hotel['reservation_status_date'], format='%d-%m-%Y')

# Extracting year and month from reservation_status_date
df_hotel['reservation_year'] = df_hotel['reservation_status_date'].dt.year
df_hotel['reservation_month'] = df_hotel['reservation_status_date'].dt.month

# Grouping data by reservation_year and reservation_month and calculating total cancellations
cancellations_by_year_month = df_hotel[df_hotel['is_canceled'] == 1].groupby(['reservation_year', 'reservation_month']).size()

# Resetting index for better plotting
cancellations_by_year_month = cancellations_by_year_month.reset_index(name='cancellation_count')

# Define month names
month_names = ['January', 'February', 'March', 'April', 'May', 'June',
               'July', 'August', 'September', 'October', 'November', 'December']

# Plotting
plt.figure(figsize=(12, 6))
sns.lineplot(data=cancellations_by_year_month, x='reservation_month', y='cancellation_count', hue='reservation_year')
plt.title('Trends in Cancellations Over Time')
plt.xlabel('Month')
plt.ylabel('Number of Cancellations')
plt.xticks(range(1, 13), month_names, rotation=45)
plt.legend(title='Year', loc='upper right')
plt.tight_layout()
plt.show()
```

**Interpretations:**
* In 2016, the month with the highest number of cancellations was September with 34 cancellations.
* 
In 2017, the month with the highest number of cancellations was March with 53 cancellations.

## Task 3.3. Can we identify high-risk bookings that are more likely to be cancelled?

```python
# Define factors influencing cancellations
factors = ['lead_time', 'distribution_channel', 'deposit_type', 'customer_type', 'previous_cancellations', 'previous_bookings_not_canceled']

# Create subplots for each factor
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(18, 10))

# Iterate over factors
for factor, ax in zip(factors, axes.flatten()):
    # Calculate cancellation rate for each category of the factor
    cancellation_rate = df_hotel.groupby(factor)['is_canceled'].mean().reset_index()
    # Plot cancellation rates
    sns.barplot(data=cancellation_rate, x=factor, y='is_canceled', ax=ax)
    ax.set_title(f'Cancellation Rate by {factor}')
    ax.set_xlabel(factor)
    ax.set_ylabel('Cancellation Rate')

# Adjust layout
plt.tight_layout()
plt.show()

# Identify high-risk bookings
high_risk_bookings = df_hotel[(df_hotel['lead_time'] > 180) | 
                               (df_hotel['distribution_channel'].isin(['Direct', 'Corporate'])) | 
                               (df_hotel['deposit_type'] == 'No Deposit') | 
                               (df_hotel['customer_type'] == 'Transient') | 
                               (df_hotel['previous_cancellations'] > 0) | 
                               (df_hotel['previous_bookings_not_canceled'] > 5)]

print("Number of High-Risk Bookings:", len(high_risk_bookings))
```

**Interpretations:**
* For lead_time, the category with the highest cancellation rate is 139 with a cancellation rate of 100 %.
* For distribution_channel, the category with the highest cancellation rate is GDS with a cancellation rate of 59 %.
* For deposit_type, the category with the highest cancellation rate is Refundable with a cancellation rate of 100 %.
* For customer_type, the category with the highest cancellation rate is Transient with a cancellation rate of 27 %.
* For previous_cancellations, the category with the highest cancellation rate is 0 with a cancellation rate of  23 %. 
* For previous_bookings_not_canceled, the category with the highest cancellation rate is 2 with a cancellation rate of 100 %

### **Task 1: Revenue Management** 
## Task 1.1.   How does the Average Daily Rate (ADR) vary over time?

```python
# Group data by arrival_date_month and calculate mean ADR for each month
adr_monthly = df_hotel.groupby('arrival_date_month')['adr'].mean()

# Define the order of months for proper plotting
months_order = ['January', 'February', 'March', 'April', 'May', 'June', 
                'July', 'August', 'September', 'October', 'November', 'December']

# Plot the ADR variation over time
plt.figure(figsize=(10, 6))
adr_monthly.reindex(months_order).plot(marker='o', color='b')
plt.title('Average Daily Rate (ADR) Variation Over Time')
plt.xlabel('Month')
plt.ylabel('Average Daily Rate (ADR)')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

**Interpretations:**
* The month(s) with the highest Average Daily Rate (ADR) are: August
* The highest ADR value is: 162.16

### Task 1.2. Can we identify pricing strategies that maximize revenue?

```python
# Calculate revenue by multiplying ADR with the number of room nights sold
df_hotel['revenue'] = df_hotel['adr'] * (df_hotel['stays_in_week_nights'] + df_hotel['stays_in_weekend_nights'])

# Grouping by ADR and calculating total revenue
revenue_by_adr = df_hotel.groupby('adr')['revenue'].sum()

# Plotting the relationship between ADR and revenue
plt.figure(figsize=(10, 6))
plt.plot(revenue_by_adr.index, revenue_by_adr.values, marker='o', linestyle='-')
plt.xlabel('Average Daily Rate (ADR)')
plt.ylabel('Total Revenue')
plt.title('Relationship between ADR and Total Revenue')
plt.grid(True)
plt.show()
```

**Interpretations:**
1. **Revenue Maximization**: Hotels can use this insight to develop pricing strategies aimed at maximizing revenue. By strategically adjusting ADR based on demand, seasonality, and other factors, hotels can optimize revenue generation while maintaining occupancy rates.
